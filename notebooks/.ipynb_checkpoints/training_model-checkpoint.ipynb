# Разбиение данных

```python
images = sorted(glob.glob('/content/drive/MyDrive/course_work/data/images/*'))
masks = sorted(glob.glob('/content/drive/MyDrive/course_work/data/masks/*'))
```

```python
images_train1, images_test, masks_train1, masks_test = train_test_split(
    images, masks, test_size=0.10, random_state=42, shuffle=True
)

images_train, images_val, masks_train, masks_val = train_test_split(
    images_train1, masks_train1, test_size=0.20, random_state=42, shuffle=True
)
```

```python
class MyDataset(Dataset):
    def __init__(self, imgs, masks, trans=None, apply_sobel=True, sobel_strength=0.5):
        self.imgs = imgs
        self.masks = masks
        self.trans = trans
        self.apply_sobel = apply_sobel
        self.sobel_strength = sobel_strength

    def __len__(self):
        return len(self.imgs)

    def __getitem__(self, i):
        img = cv2.imread(self.imgs[i])
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        if self.apply_sobel:
            img = sobel_enhance(img, strength=self.sobel_strength)

        mask = cv2.imread(self.masks[i], cv2.IMREAD_GRAYSCALE)
        mask = (mask > 0).astype(np.float32)

        if self.trans:
            a = self.trans(image=img, mask=mask)
            img, mask = a['image'], a['mask']

        mask = np.expand_dims(mask, 0)
        return img, torch.tensor(mask, dtype=torch.float32)
```

# Усиление границ с помощью оператора Собеля

```python
def sobel_enhance(img, strength=0.5):
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY).astype(np.float32)

    Kx = np.array([[-1, 0, 1],
                   [-2, 0, 2],
                   [-1, 0, 1]], dtype=np.float32)
    Ky = np.array([[-1, -2, -1],
                   [ 0,  0,  0],
                   [ 1,  2,  1]], dtype=np.float32)

    gx = cv2.filter2D(gray, -1, Kx)
    gy = cv2.filter2D(gray, -1, Ky)

    mag = np.sqrt(gx * gx + gy * gy)
    mmax = mag.max()
    if mmax > 0:
        mag = mag / mmax
    else:
        mag = np.zeros_like(mag)
    mag_u8 = (np.clip(mag * 255.0, 0, 255)).astype(np.uint8)

    enhanced = img.astype(np.float32)

    enhanced[:, :, 0] = np.clip(enhanced[:, :, 0] + strength * 0.5 * mag_u8, 0, 255)
    enhanced[:, :, 1] = np.clip(enhanced[:, :, 1] + strength * 0.4 * mag_u8, 0, 255)
    enhanced[:, :, 2] = np.clip(enhanced[:, :, 2] + strength * 0.6 * mag_u8, 0, 255)

    return enhanced.astype(np.uint8)
```

# Вспомогательные функции(ошибки, преобразование изображения)

```python
def get_trans(phase='train'):
    if phase == 'train':
        return A.Compose([
            A.Resize(512, 512),
            A.HorizontalFlip(p=0.5),
            A.Normalize(),
            ToTensorV2()
        ])
    else:
        return A.Compose([
            A.Resize(512, 512),
            A.Normalize(),
            ToTensorV2()
        ])
```

```python
def iou(pred, gt):
    pred = (pred > 0.5).float()
    gt = (gt > 0.5).float()
    inter = (pred * gt).sum()
    union = pred.sum() + gt.sum() - inter
    return inter / union
```

```python
def dice_coef(pred, target, eps=1e-6):
    pred = (pred > 0.5).float()
    target = target.float()

    inter = (pred * target).sum()
    union = pred.sum() + target.sum()

    return (2 * inter + eps) / (union + eps)
```

```python
def dice(logits, target, eps=1e-6):
    probs = torch.sigmoid(logits)
    inter = (probs * target).sum()
    union = probs.sum() + target.sum()
    return 1 - (2 * inter + eps) / (union + eps)
```

```python
class BCEDiceLoss(nn.Module):
    def __init__(self):
        super().__init__()
        self.bce = nn.BCEWithLogitsLoss()

    def forward(self, logits, target):
        return self.bce(logits, target) + dice(logits, target)
```

# Построение U-Net


```python
class Encoder(nn.Module):
    def __init__(self):
        super().__init__()
        rn = models.resnet101(weights=models.ResNet101_Weights.DEFAULT)
        self.init = nn.Sequential(rn.conv1, rn.bn1, rn.relu)
        self.pool = rn.maxpool
        self.l1 = rn.layer1
        self.l2 = rn.layer2
        self.l3 = rn.layer3
        self.l4 = rn.layer4

    def forward(self, x):
        x0 = self.init(x)
        x1 = self.l1(self.pool(x0))
        x2 = self.l2(x1)
        x3 = self.l3(x2)
        x4 = self.l4(x3)
        return x0, x1, x2, x3, x4
```

```python
class Up(nn.Module):
    def __init__(self, in_ch, skip_ch, out_ch):
        super().__init__()
        self.up = nn.ConvTranspose2d(in_ch, out_ch, 2, 2)
        self.conv = nn.Sequential(
            nn.Conv2d(out_ch + skip_ch, out_ch, 3, 1, 1),
            nn.ReLU(),
            nn.Conv2d(out_ch, out_ch, 3, 1, 1),
            nn.ReLU()
        )

    def forward(self, x, skip):
        x = self.up(x)
        dx = skip.size(3) - x.size(3)
        dy = skip.size(2) - x.size(2)
        x = F.pad(x, [dx // 2, dx - dx // 2, dy // 2, dy - dy // 2])
        x = torch.cat([skip, x], 1)
        return self.conv(x)
```

```python
class MyUNet(nn.Module):
    def __init__(self):
        super().__init__()
        self.enc = Encoder()
        self.u1 = Up(2048, 1024, 1024)
        self.u2 = Up(1024, 512, 512)
        self.u3 = Up(512, 256, 256)
        self.u4 = Up(256, 64, 64)
        self.final = nn.ConvTranspose2d(64, 64, 2, 2)
        self.outc = nn.Conv2d(64, 1, 1)

    def forward(self, x):
        x0, x1, x2, x3, x4 = self.enc(x)
        x = self.u1(x4, x3)
        x = self.u2(x, x2)
        x = self.u3(x, x1)
        x = self.u4(x, x0)
        x = self.final(x)
        return self.outc(x)
```

# Обработка данных

```python
train_dataset = MyDataset(
    images_train,
    masks_train,
    trans=get_trans('train'),
    apply_sobel=True,
    sobel_strength=0.5
)

val_dataset = MyDataset(
    images_val,
    masks_val,
    trans=get_trans('val'),
    apply_sobel=False
)

test_dataset = MyDataset(
    images_test,
    masks_test,
    trans=get_trans('test'),
    apply_sobel=False
)

train_loader = DataLoader(
    train_dataset,
    batch_size=4,
    shuffle=True,
    num_workers=2,
    pin_memory=True
)

val_loader = DataLoader(
    val_dataset,
    batch_size=4,
    shuffle=False,
    num_workers=2,
    pin_memory=True
)

test_loader = DataLoader(
    test_dataset,
    batch_size=4,
    shuffle=False,
    num_workers=2,
    pin_memory=True
)
```

# Обучение

```python
from torch.cuda.amp import autocast, GradScaler
```

```python
BATCH_SIZE = 64
NUM_WORKERS = 16
PIN_MEMORY = True
```

```python
train_loader = DataLoader(
    train_dataset,
    batch_size=BATCH_SIZE,
    shuffle=True,
    num_workers=NUM_WORKERS,
    pin_memory=PIN_MEMORY
)
```

```python
val_loader = DataLoader(
    val_dataset,
    batch_size=BATCH_SIZE,
    shuffle=False,
    num_workers=NUM_WORKERS,
    pin_memory=PIN_MEMORY
)
```

```python
model = MyUNet().to(device)
opt = torch.optim.Adam(model.parameters(), lr=1e-4)
loss_fn = BCEDiceLoss()
```

```python
scaler = GradScaler()

history = {'train_loss': [],
           'val_loss': [],
           'val_iou': [],
           'val_dice': []}

best_iou = 0
cnt = 0
NUM_EPOCHS = 20
```

```python
torch.backends.cudnn.benchmark = True
```

```python
def train_one_epoch(model, loader, optimizer, scaler, loss_fn, device):
    model.train()
    running_loss = 0.0
    pbar = tqdm(loader, desc='Train')

    for img, msk in pbar:
        img, msk = img.to(device, non_blocking=True), msk.to(device, non_blocking=True)

        optimizer.zero_grad()
        with autocast(dtype=torch.float16):
            logits = model(img)
            loss = loss_fn(logits, msk)

        scaler.scale(loss).backward()
        scaler.step(optimizer)
        scaler.update()

        running_loss += loss.item() * img.size(0)
        pbar.set_postfix({'loss': running_loss / ((pbar.n + 1) * loader.batch_size)})

    epoch_loss = running_loss / len(loader.dataset)
    return epoch_loss
```

```python
def validate_one_epoch(model, loader, loss_fn, device):
    model.eval()
    val_losses = 0.0
    ious = []
    dices = []

    with torch.no_grad():
        pbar = tqdm(loader, desc='Val')
        for img, msk in pbar:
            img, msk = img.to(device, non_blocking=True), msk.to(device, non_blocking=True)

            with autocast(dtype=torch.float16):
                logits = model(img)
                loss = loss_fn(logits, msk)
                pr = torch.sigmoid(logits)

            val_losses += loss.item() * img.size(0)
            ious.append(iou(pr.float(), msk.float()).item())
            dices.append(dice_coef(pr.float(), msk.float()).item())

            pbar.set_postfix({
                'val_loss': val_losses / ((pbar.n + 1) * loader.batch_size),
                'iou': np.mean(ious),
                'dice': np.mean(dices)
            })

    epoch_val_loss = val_losses / len(loader.dataset)
    epoch_val_iou = np.mean(ious)
    epoch_val_dice = np.mean(dices)

    return epoch_val_loss, epoch_val_iou, epoch_val_dice
```

```python
for i in range(NUM_EPOCHS):
    print(f'Epoch {i+1}/{NUM_EPOCHS}')

    epoch_train_loss = train_one_epoch(model, train_loader, opt, scaler, loss_fn, device)
    history['train_loss'].append(epoch_train_loss)

    model.eval()
    val_losses = 0.0
    ious = []
    dices = []

    with torch.no_grad():
        pbar = tqdm(val_loader, desc='Val')
        for img, msk in pbar:
            img, msk = img.to(device, non_blocking=True), msk.to(device, non_blocking=True)

            with torch.autocast(device_type='cuda', dtype=torch.float16):
                logits = model(img)
                loss = loss_fn(logits, msk)
                pr = torch.sigmoid(logits)

            val_losses += loss.item() * img.size(0)
            ious.append(iou(pr.float(), msk.float()).item())
            dices.append(dice_coef(pr.float(), msk.float()).item())

            pbar.set_postfix({
                'val_loss': val_losses / ((pbar.n + 1) * val_loader.batch_size),
                'iou': np.mean(ious),
                'dice': np.mean(dices)
            })

    epoch_val_loss = val_losses / len(val_loader.dataset)
    epoch_val_iou = np.mean(ious)
    epoch_val_dice = np.mean(dices)

    history['val_loss'].append(epoch_val_loss)
    history['val_iou'].append(epoch_val_iou)
    history['val_dice'].append(epoch_val_dice)

    print(f'Epoch {i+1} summary: val_loss={epoch_val_loss:.4f}, IoU={epoch_val_iou:.4f}, Dice={epoch_val_dice:.4f}')

    if epoch_val_iou > best_iou:
        best_iou = epoch_val_iou
        save_path = os.path.join('/content/drive/MyDrive/course_work/models', 'best_model.pth')
        torch.save(model.state_dict(), save_path)
        cnt = 0
    else:
        cnt += 1

    if cnt >= 5:
        break
```

```python
model.eval()

all_preds = []
all_gts = []

with torch.no_grad():
    for img, msk in tqdm(val_loader, desc='Collect PR data'):
        img = img.to(device)
        msk = msk.to(device)

        logits = model(img)
        probs = torch.sigmoid(logits)

        all_preds.append(probs.cpu().numpy().ravel())
        all_gts.append(msk.cpu().numpy().ravel())

all_preds = np.concatenate(all_preds)
all_gts = np.concatenate(all_gts)
```

```python
thresholds = np.linspace(0.0, 1.0, 101)

precisions = []
recalls = []

for t in thresholds:
    preds_bin = (all_preds >= t).astype(np.uint8)

    tp = np.sum((preds_bin == 1) & (all_gts == 1))
    fp = np.sum((preds_bin == 1) & (all_gts == 0))
    fn = np.sum((preds_bin == 0) & (all_gts == 1))

    precision = tp / (tp + fp + 1e-6)
    recall = tp / (tp + fn + 1e-6)

    precisions.append(precision)
    recalls.append(recall)
```

```python
plt.figure(figsize=(7, 6))
plt.plot(recalls, precisions)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision–Recall кривая')
plt.grid(True)
plt.show()
```

```python
model.eval()

test_ious = []
test_dices = []
```

```python
with torch.no_grad():
    for img, gt in tqdm(test_loader, desc='Test metrics'):
        img = img.to(device)
        gt = gt.to(device)

        logits = model(img)
        probs = torch.sigmoid(logits)
        pred = (probs > 0.5).float()

        test_ious.append(iou(pred.cpu(), gt.cpu()).item())
        test_dices.append(dice_coef(pred.cpu(), gt.cpu()).item())
```
